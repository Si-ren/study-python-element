```python
import pandas as pd
import numpy as np
```

```python
# 1. 用三种不同的方法，创建以下Dataframe（保证columns和index一致，值不做要求）

# 结果Dataframe为：
#     four  one  three  two
# a     4    1      3    2
# b     5    2      4    3
# c     6    3      5    4
# d     7    4      6    5
# e     8    5      7    6
```

```python

data = [
    [4,1,3,2],
    [5,2,4,3],
    [6,3,5,4],
    [7,4,6,5],
    [8,5,7,6]
]
df = pd.DataFrame(data,index=["a","b","c","d","e"],columns=["four","one","three","two"])
df
```

```python
data = {
    "four":[4,5,6,7,8],
    "one":[1,2,3,4,5],
    "three":[3,4,5,6,7],
    "two":[2,3,4,5,6],
}
df = pd.DataFrame(data,index=["a","b","c","d","e"])
df
```

```python
data = [
    {"four":4,"one":1,"three":3,"two":2},
    {"four":5,"one":2,"three":4,"two":3},
    {"four":6,"one":3,"three":5,"two":4},
    {"four":7,"one":4,"three":6,"two":5},
    {"four":8,"one":5,"three":7,"two":6},
]
df = pd.DataFrame(data,index=["a","b","c","d","e"])
df
```

```python
data = {
    'four' : pd.Series([4,5,6,7,8], index=["a","b","c","d","e"]),
    'one' : pd.Series([1,2,3,4,5], index=["a","b","c","d","e"]),
    'three' : pd.Series([3,4,5,6,7], index=["a","b","c","d","e"]),
    'two' : pd.Series([2,3,4,5,6], index=["a","b","c","d","e"]),
  }
df = pd.DataFrame(data)
df
```

```python
# 2. 如图创建Dataframe(4*4，值为0-100的随机数)，通过索引得到以下值
#     ① 索引得到b，c列的所有值
#     ② 索引得到第三第四行的数据
#     ③ 按顺序索引得到two，one行的值
#     ④ 索引得到大于50的值
#           a          b          c          d
# one    60.936882  34.198569  86.933961  63.217850
# two    93.910622   8.843498  12.482240  35.940462
# three  11.350391  40.704308  50.524502   5.215897
# four    5.777448  75.515444  96.847913  57.683561
```

```python
# 创建列表
# 这里使用了np.reshape函数把一维数组变为二维
df = pd.DataFrame(np.reshape(np.random.rand(16)*100,(4,4)) ,index=["one","two","three","four"],columns=["a","b","c","d"])
df
```

```python
# 索引得到b,c列的值
b = df["b"].values
print(b,type(b))
c = df["c"]
print(c,type(c))
```

```python
# 索引得到第三第四行的数据
# loc函数和iloc函数参数为[行,列]
# print(df.loc['two'])
print(df.iloc[2],type(df.iloc[2]))
print(df.iloc[3],type(df.iloc[3]))
```

```python
#  索引得到大于50的值
result = df[df>50]
print(result,type(result))
```

```python
# 按照index降序
# 这是算着每行的总和进行降序
# - axis：0按照行名排序；1按照列名排序
# - ascending：默认True升序排列；False降序排列
# - inplace：默认False，否则排序之后的数据直接替换原来的数据
df.sort_index(axis=0, ascending=False)
df.sort_index(axis=1, ascending=False)
```

```python
#按照列降序
#先按b列降序，再按a列升序排序
df.sort_values(by=['b','a'],ascending=[False,True]) 
```
